```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
header = ["age", "workclass", "someid","education", "education_num",
          "marital_status", "occupation", "relationship", "race", "sex", "capital_gain","capital_loss",
          "hours_per_week", "native_country", "class"]
df = pd.read_csv("adults.csv", header = None, names = header, na_values = [' ?'])
```

```python
df=  df.drop_duplicates()
```

```python
df=  df.drop_duplicates(inplace=True) # this is another syntax to drop duplicates value
```

```python
df.workclass.unique()
```

```python
df.isna().sum() #another syntax for find the null values
```

```python
# if we have null values in string or text datatype we can not use mean , median
#intead of that we can use mode >> it will give maximum frequency
```

```python
cols = ['workclass', 'occupation', 'native_country']
for col in cols:
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
df.native_country.unique()
```

```python
for col in [col for col in df.columns if df[col].dtype == 'object']:
  df[col] = df[col].str.strip() # strip remove all extra and whitespaces
```

```python
df.native_country.unique()
```

```python
df.dtypes
```

```python
# model always wants numeric value it will not take text
# we use labelencoder to covert categorical data to number so we cn use in modeling
#onehotencoding will create of each value one column so while we have so many values in one col
# use labelencoding
```

```python
from sklearn.preprocessing import LabelEncoder

Label = LabelEncoder()
for col in [col for col in df.columns if df[col].dtype == 'object']:
  df[col] = Label.fit_transform(df[col])
```

```python
df
```

```python
df.dtypes
```

```python
y = df['class']
```

```python
y
```

```python
X = df.drop("class", axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
#we fit dataset on train data
#we transfrom on test data
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from xgboost import XGBClassifier # this is highest so we wil take this algorithm
clf = XGBClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import AdaBoostClassifier
clf = AdaBoostClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy_score(y_test, y_pred)
```
