```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#Q1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.

#ans >> 
sns.get_dataset_names()
```

```python
np.random.seed(42)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 values from 0 to 10
y = 2 * x + np.random.normal(0, 1, 100)  # Linear relationship with noise

# Step 2: Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.7, edgecolors='w', s=100)
plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.grid(True)
plt.show()
```

```python
#Q2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

#ans>>
```

```python
np.random.seed(42)  
data = np.random.normal(loc=0, scale=1, size=1000)  


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')

plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.kdeplot(data, fill=True, color='blue', alpha=0.6)
plt.title('Density Plot of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Density')
```

```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
```

```python
#ans >>
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': [23, 45, 56, 78, 33]
}

# Convert the dataset into a pandas DataFrame
df = pd.DataFrame(data)

# Step 2: Visualize the comparison using a bar plot
plt.figure(figsize=(10, 6))

# Create a bar plot with seaborn
sns.barplot(x='Category', y='Value', data=df, palette='viridis')

# Add title and labels
plt.title('Comparison of Categories Based on Values')

plt.xlabel('Category')
plt.ylabel('Value')
```

```python
#Q4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
```

```python
#ans>> 
np.random.seed(42)  # For reproducibility
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, size=400),
    'Value': np.concatenate([np.random.normal(loc=i*10, scale=5, size=100) for i in range(len(categories))])
}


df = pd.DataFrame(data)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='coolwarm')
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
#Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
#ans >>
np.random.seed(42)  

# Number of samples
n_samples = 1000

# Generate random data
x1 = np.random.normal(0, 1, n_samples)
x2 = 0.5 * x1 + np.random.normal(0, 1, n_samples)  # Correlated with x1
x3 = -0.3 * x1 + 0.2 * x2 + np.random.normal(0, 1, n_samples)  # Correlated with x1 and x2

# Create a DataFrame
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3
})

# Step 2: Compute the correlation matrix
corr_matrix = df.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
