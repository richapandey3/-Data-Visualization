```python
np.random.seed(30)

# Generate dataset
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract features for the 3D scatter plot
x = df['X']
y = df['Y']
z = df['Z']

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot data points
scatter = ax.scatter(x, y, z, c='blue', marker='o')

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Add a title
plt.title('3D Scatter Plot of Uniformly Distributed Data')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(15)

# Generate synthetic dataset
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}

# Create DataFrame
df = pd.DataFrame(data)

# Create a violin plot to display the distribution of scores across different grades
plt.figure(figsize=(12, 8))
sns.violinplot(x='Grade', y='Score', data=df, palette='muted')

# Add title and labels
plt.title('Distribution of Scores Across Different Grade Categories')
plt.xlabel('Grade')
plt.ylabel('Score')

# Show the plot
plt.show()
```

```python

# Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# Seed for reproducibility
np.random.seed(20)

# Generate synthetic sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the data for heatmap
pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap to visualize the variation in sales
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=.5)

# Add title and labels
plt.title('Heatmap of Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

# Show the plot
plt.show()
```

```python
#Q4. Using the given x and y data, generate a 3D surface plot to visualize the function
```

```python
#ans>>
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)


# Create a 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
surf = ax.plot_surface(df['X'].values.reshape(100, 100), 
                       df['Y'].values.reshape(100, 100), 
                       df['Z'].values.reshape(100, 100), 
                       cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of sin(sqrt(X^2 + Y^2))')

# Show the plot
plt.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
plt.show()
```

```python
#Q5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
```

```python

# Seed for reproducibility
np.random.seed(25)

# Generate synthetic dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Bubble chart
plt.figure(figsize=(12, 8))

# Plot bubbles with GDP on x-axis, Population on y-axis, and size proportional to Population
scatter = plt.scatter(
    df['GDP'], 
    df['Population'], 
    s=df['Population'],  # Bubble size proportional to Population
    alpha=0.6,           # Transparency
    c=df['Population'], # Color by Population
    cmap='viridis',      # Colormap
    edgecolors='w',     # White edge color for bubbles
    linewidth=0.5       # Edge width
)

# Add labels and title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: GDP vs Population with Bubble Size Proportional to Population')

# Add a colorbar for the population
plt.colorbar(scatter, label='Population')

# Add annotations for each country
for i in range(len(df)):
    plt.text(df['GDP'].iloc[i], df['Population'].iloc[i], df['Country'].iloc[i], fontsize=12, ha='right')

# Show the plot
plt.grid(True)
plt.show()
```
