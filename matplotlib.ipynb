```python
#Q3. Display a bar chart to represent the frequency of each item in the given array categories.
```

```python
#Ans>> 

categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 35, 20]
```

```python
plt.bar(categories, values, color = 'red')
plt.title("bar chart")
plt.xlabel("categories")
plt.ylabel("values")
plt.show()

```

```python
#4. Create a histogram to visualize the distribution of values in the array data.
```

```python
#Ans>> 

data = np.random.normal(0, 1, 1000)

plt.hist(data, color = 'r')
plt.title("histogram")
plt.show()   
```

```python
#Q5. Show a pie chart to represent the percentage distribution of different sections in the array `sections?
```

```python
#Ans>>

sections = ['Section A', 'Section B', 'Section C', 'Section D']
sizes = [25, 30, 15, 30]

plt.pie(sizes, labels = sections, autopct="%1.1f%%")
plt.title("pie chart")
plt.show()
```
