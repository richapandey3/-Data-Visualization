```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with the data
p.line(x, y, legend_label="y=sin(x)", line_width=2)

# Show the results
show(p)
```

```python
#Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for themarkers based on the 'sizes' and 'colors' columns.
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

output_notebook()

# Generate random data
np.random.seed(42)
N = 100
x = np.random.rand(N) * 100
y = np.random.rand(N) * 100
sizes = np.random.rand(N) * 50 + 10  # Sizes between 10 and 60
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], N)

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create the scatter plot
p = figure(title="Bokeh Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)
```

```python
#Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.transform import dodge
from bokeh.io import show

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh plot
output_file("fruit_counts.html")

p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
#Q4. Create a Bokeh histogram to visualize the distribution of the given data
```

```python
#ans >>
# Generate data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh plot
output_file("histogram.html")

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color='navy', line_color='white', alpha=0.5)

# Show the plot
show(p)
```

```python
#Q5. Create a Bokeh heatmap using the provided dataset
```

```python
#ans>>

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.layouts import layout
from bokeh.models import ColumnDataSource

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for use in Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Define the output file
output_file("heatmap.html")

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           tooltips=[("Value", "@value")])

# Define a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=min(data_flat), high=max(data_flat))

# Add the heatmap squares
p.rect(x="x", y="y", width=1/10, height=1/10, source=source,
       line_color=None, fill_color=transform('value', mapper))

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
